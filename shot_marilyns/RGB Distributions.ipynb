- PILLOW is one optional module to read digital images
- For PNG file: RGBA means Red, Green, Blue, Alpha where Alpha is transparency.

```python
# Code for RGB Distributions

from PIL import Image
from numpy import asarray
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Function

def rgbdist(image):
    
    array = asarray(image)
        
    image2 = Image.fromarray(array)
    
    red = []
    
    green = []
    
    blue = []
    
    width, height = image2.size
    
    for y in range(height):
        
        for x in range(width):
            
            # Get the RGB values of the pixel
            
            try:
                (r, g, b, a) = image2.getpixel((x, y))
                
            except:    
                (r, g, b) = image2.getpixel((x,y))

            # Append the RGB values to the corresponding lists
            red.append(r)
            green.append(g)
            blue.append(b)
        
    sns.kdeplot(data=red, color='red', shade=False)
    sns.kdeplot(data=green, color='green', shade=False)
    sns.kdeplot(data=blue, color='blue', shade=False)

    plt.xlabel('Channel')
    plt.ylabel('Density')
    plt.show
```

```python
marilyn_blue = Image.open('shot_blue_marilyn_1964-19604.png')
marilyn_eggblue = Image.open('shot_eggblue_marilyn_1964-19604.png')
marilyn_orange = Image.open('shot_orange_marilyn_1964-19604.png')
marilyn_red = Image.open('shot_red_marilyn_1964-19604.png')
marilyn_turq = Image.open('shot_turq_marilyn_1964-19604.png')
```

```python
rgbdist(marilyn_blue)
plt.title('Blue Marilyn RGB Distributions')
```

```python
rgbdist(marilyn_eggblue)
plt.title('Egg Blue Marilyn RGB Distributions')
```

```python
rgbdist(marilyn_orange)
plt.title('Orange Marilyn RGB Distributions')
```

```python
rgbdist(marilyn_red)
plt.title('Red Marilyn RGB Distributions')
```

```python
rgbdist(marilyn_turq)
plt.title('Turquoise Marilyn RGB Distributions')
```
