```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')
data.head()
```

```python
data.info()
```

### Visualizations

```python
def show_boxplot(data):
    plt.rcParams['figure.figsize'] = [14,6]
    sns.boxplot(data = data, orient="v")
    plt.title("Distribution of Variables", fontsize = 16)
    plt.ylabel("Range", fontweight = 'bold')
    plt.xlabel("Attributes", fontweight = 'bold')
    plt.xticks(rotation=90)
   
show_boxplot(data)
```

```python
def remove_outliers(data):
   
    df = data.copy()
   
    for col in list(df.columns):
 
        Q1 = df[str(col)].quantile(0.05)
        Q3 = df[str(col)].quantile(0.95)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5*IQR
        upper_bound = Q3 + 1.5*IQR
 
        df = df[(df[str(col)] >= lower_bound) & (df[str(col)] <= upper_bound)]
   
    return df
```

```python
without_outliers = remove_outliers(data)
```

```python
show_boxplot(without_outliers)
```
